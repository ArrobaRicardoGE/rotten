```python
import pandas as pd
import numpy as np
```

```python
critics = pd.read_csv('dataset/f_critics.csv', index_col=0)
audience = pd.read_csv('dataset/f_audience.csv', index_col=0)
reviews_aa = pd.read_csv('dataset/audience_fresh.csv', index_col=0)
reviews_ac = pd.read_csv('dataset/audience_rotten.csv', index_col=0)
reviews_ca = pd.read_csv('dataset/critics_rotten.csv', index_col=0)
reviews_cc = pd.read_csv('dataset/critics_fresh.csv', index_col=0)
```

```python
reviews_aa.dropna(inplace=True)
reviews_ac.dropna(inplace=True)
reviews_ca.dropna(inplace=True)
reviews_cc.dropna(inplace=True)
```

```python
audience_reviews = pd.concat([reviews_aa, reviews_ac])
critics_reviews = pd.concat([reviews_ca, reviews_cc])
```

```python
movie_ids = pd.read_csv('dataset/movie_ids.csv', index_col=0).set_index('id')
```

```python
audience_reviews = audience_reviews.merge(movie_ids, left_index=True, right_index=True).set_index('rotten_tomatoes_link')
audience_reviews
```

```python
critics_reviews = critics_reviews.merge(movie_ids, left_index=True, right_index=True).set_index('rotten_tomatoes_link')
critics_reviews
```

```python
audience_reviews.review_content = audience_reviews.review_content.str.lower()
audience_reviews
```

```python
critics_reviews.review_content = critics_reviews.review_content.str.lower()
critics_reviews
```

```python
critics_reviews = critics_reviews.drop(columns=['is_fresh'])
critics_reviews
```

```python
audience_reviews = audience_reviews.drop(columns=['is_fresh'])
audience_reviews
```

```python
critics_reviews.to_csv('dataset/critics_reviews_clean_v2.csv')
audience_reviews.to_csv('dataset/audience_reviews_clean_v2.csv')
```
